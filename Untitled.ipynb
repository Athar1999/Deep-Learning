```python
## 1. Imports and Data Loading
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load and Preprocess
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

# Reusable Training Function
def build_and_train(activation='relu', optimizer='adam', loss='sparse_categorical_crossentropy'):
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation=activation),
        Dense(64, activation=activation),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    history = model.fit(x_train, y_train if loss=='sparse_categorical_crossentropy' else y_train_cat, 
                        epochs=5, validation_split=0.2, verbose=0)
    return model, history

## 2. Experiments: Activation Functions
activations = ['relu', 'sigmoid', 'tanh']
act_histories = {}

for act in activations:
    print(f"Training with {act}...")
    _, hist = build_and_train(activation=act)
    act_histories[act] = hist

# Graphs for Activations
plt.figure(figsize=(12, 4))
for act in activations:
    plt.plot(act_histories[act].history['val_accuracy'], label=act)
plt.title('Activation Functions Comparison (Accuracy)')
plt.legend()
plt.show()

## 3. Experiments: Optimizers
optimizers = ['adam', 'sgd']
opt_histories = {}

for opt in optimizers:
    print(f"Training with {opt}...")
    _, hist = build_and_train(optimizer=opt)
    opt_histories[opt] = hist

# Graphs for Optimizers
plt.figure(figsize=(12, 4))
for opt in optimizers:
    plt.plot(opt_histories[opt].history['val_loss'], label=opt)
plt.title('Optimizers Comparison (Loss)')
plt.legend()
plt.show()

## 4. Experiments: Cost Functions
# Sparse Categorical vs Categorical Crossentropy
_, h1 = build_and_train(loss='sparse_categorical_crossentropy')
_, h2 = build_and_train(loss='categorical_crossentropy')

plt.plot(h1.history['val_loss'], label='Sparse CE')
plt.plot(h2.history['val_loss'], label='Categorical CE')
plt.title('Loss Function Impact')
plt.legend()
plt.show()
```
